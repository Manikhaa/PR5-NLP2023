```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df_train = pd.read_csv('train.csv')
```

```python
print(df_train.head())
```

```python
# (a) Distribusi Sales Price
plt.figure(figsize=(10, 6))
sns.histplot(df_train['SalePrice'], kde=True)
plt.title('Distribusi Sales Price')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# (b) Scatter plot untuk sales price vs PoolArea
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PoolArea', y='SalePrice', data=df_train)
plt.title('Scatter plot untuk Sales Price vs PoolArea')
plt.xlabel('Pool Area')
plt.ylabel('Sale Price')
plt.show()
```

```python
# (c) Box-Plot untuk Sales price vs GarageQual
plt.figure(figsize=(12, 8))
sns.boxplot(x='GarageQual', y='SalePrice', data=df_train)
plt.title('Box-Plot untuk Sales Price vs GarageQual')
plt.xlabel('Garage Quality')
plt.ylabel('Sale Price')
plt.show()
```

```python
# (d) Correlation matrix
correlation_matrix = df_train.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Analisis korelasi
top_correlations = correlation_matrix['SalePrice'].sort_values(ascending=False)[1:6]
print('Top 5 variabel yang paling mempengaruhi harga rumah:')
print(top_correlations)
```
